### Business Question 1: Code:

```python
# Function to Retrieve Company Details from API

def _get_company_details(api_key, sic_codes, location="Edinburgh", page=1, items_per_page=20):
    url = "https://api.company-information.service.gov.uk/advanced-search/companies"
    headers = {"Authorization": api_key}
    params = {"sic_codes": sic_codes, "start_index": (page - 1) * items_per_page, "items_per_page": items_per_page,
              "location": location}
    
    try:
        response = requests.get(url, headers=headers, params=params)
        # Raises an HTTPError for bad responses
        response.raise_for_status()  

        return response.json()
    except requests.exceptions.RequestException:
        # Return None without printing the error
        return None
```

```python
# Retrieves company details based on SIC codes with pagination

def get_company_details(sic_codes, api_key='a3bd779a-8a23-405f-9a9c-3c7351288b16', location="Edinburgh",
                        items_per_page=20, total_items=2000):

    # Initialize an empty list to store all items
    all_items = []

    # Retrieve data in chunks until the desired number of items is reached
    current_page = 1
    while len(all_items) < total_items:
        result = _get_company_details(api_key, sic_codes, location=location, page=current_page, items_per_page=items_per_page)
        if not result or not result.get('items'):
            break

        all_items.extend(result['items'])
        current_page += 1

    # Create a DataFrame from all retrieved items
    return pd.DataFrame(all_items)
```

```python
# Function to Extract and Count Companies Created per Year (2013-2023)

def companies_created(df):
    # Convert 'date_of_creation' to datetime
    df = df.assign(date_of_creation=pd.to_datetime(df['date_of_creation'], errors='coerce'))

    # Filter rows for the years 2013-2023
    df_filtered = df[df['date_of_creation'].dt.year.between(2013, 2023)].copy()

    # Extract the year and create new columns
    df_filtered = df_filtered.assign(year_of_creation=df_filtered['date_of_creation'].dt.year)

    # Count dissolved companies by year
    filtered_created_counts = df_filtered.groupby('year_of_creation')['company_number'].count()

    return dict(filtered_created_counts)
```

```python
# Extract and Count Dissolved Companies by Year

def companies_dissolved(df):
    # Convert 'date_of_cessation' and 'date_of_creation' to datetime
    df = df.assign(date_of_cessation=pd.to_datetime(df['date_of_cessation'], errors='coerce'))

    # Filter rows for the years 2013-2023
    df_filtered = df[df['date_of_cessation'].dt.year.between(2013, 2023)].copy()

    # Extract the year and create new columns
    df_filtered = df_filtered.assign(year_of_cessation=df_filtered['date_of_cessation'].dt.year)

    # Filter rows for dissolved companies
    filtered_dissolved = df_filtered[df_filtered['company_status'] == 'dissolved'].copy()

    # Count dissolved companies by year
    dissolved_counts = filtered_dissolved.groupby('year_of_cessation')['company_number'].count()

    return dict(dissolved_counts)
```

```python
# Function to Calculate Created-to-Dissolved Ratio

def dissolved_created_ratio(dict1,dict2,decimal_points=2):
    result_dict = {key: round(dict1[key] / dict2[key], decimal_points) for key in dict1}
    return result_dict
```

```python
# Analyzing Created and Dissolved Ratios for Licensed Restaurants Companies

# Fetching company details for businesses with SIC code 56101 (Licensed Restaurants)
df_licencedRestaurants = get_company_details(sic_codes="56101")

# Calculating the number of dissolved companies among licensed restaurants
licencedRestaurants_dissolved = companies_dissolved(df_licencedRestaurants)

# Calculating the number of newly created companies among licensed restaurants
licencedRestaurants_created = companies_created(df_licencedRestaurants)

# Calculating the ratio of created to dissolved companies among licensed restaurants
licencedRestaurants_ratio = dissolved_created_ratio(licencedRestaurants_created, licencedRestaurants_dissolved)

# Printing the resulting ratio
print(licencedRestaurants_ratio)
```

```python
# Analyzing Created and Dissolved Ratios for Unlicensed restaurants and cafes

# Fetching company details for businesses with SIC code 56102 (Unlicensed restaurants and cafes)
df_Unlicensed_restaurants = get_company_details(sic_codes="56102")

# Calculating the number of dissolved companies among Unlicensed restaurants and cafes
Unlicensed_restaurants_dissolved = companies_dissolved(df_Unlicensed_restaurants)

# Calculating the number of newly created companies among unlicensed restaurants and cafes
Unlicensed_restaurants_created = companies_created(df_Unlicensed_restaurants)

# Calculating the ratio of created to dissolved companies
unlicencedRestaurants_ratio = dissolved_created_ratio(Unlicensed_restaurants_created,Unlicensed_restaurants_dissolved)

# Printing the resulting ratio
print(unlicencedRestaurants_ratio)
```

```python
# Analyzing Created and Dissolved Ratios for Other food services

# Fetching company details for businesses with SIC code 56290 (Other food services)
df_otherfood = get_company_details(sic_codes="56290") 

# Calculating the number of dissolved companies
otherfood_dissolved = companies_dissolved(df_otherfood)

# Calculating the number of newly created companies
otherfood_created = companies_created(df_otherfood)

# Calculating the ratio of created to dissolved companies
otherfood_ratio = dissolved_created_ratio(otherfood_created,otherfood_dissolved)

# Printing the resulting ratio
print(otherfood_ratio)
```

```python
# Function to Retrieve Vegetable Company Details in the UK

def _vegetable_company_details_UK(api_key, sic_codes, page=1, items_per_page=20):
    
    url = "https://api.company-information.service.gov.uk/advanced-search/companies"
    headers = {"Authorization": api_key}
    params = {"sic_codes": sic_codes, "start_index": (page - 1) * items_per_page, "items_per_page": items_per_page}
    
    try:
        response = requests.get(url, headers=headers, params=params)
        response.raise_for_status()  # Raises an HTTPError for bad responses

        df_active = pd.DataFrame(response.json().get('items', []))

        # Custom function to extract locality
        def extract_locality(address):
            return address.get('locality', None)

        # Use .loc to set values in the DataFrame
        df_active.loc[:, 'locality'] = df_active['registered_office_address'].apply(extract_locality)

        return df_active.copy()  
    except requests.exceptions.RequestException as e:
        #print(f"Exception Message: {e}")
        return None
    
```

```python
# Function to Retrieve Vegetable Company Details in the UK

def vegetable_company_details_UK(sic_codes, api_key='a3bd779a-8a23-405f-9a9c-3c7351288b16',
                                 items_per_page=20, total_items=10000):
    # Initialize an empty list to store data
    all_items = []
    
    # Set the initial page number
    current_page = 1
    
    # Continue fetching data until the target number of items is reached
    while len(all_items) < total_items:
        # Call the internal function to get data for the current page
        result = _vegetable_company_details_UK(api_key, sic_codes, page=current_page, items_per_page=items_per_page)
        
        # Check if the result is None or empty, exit the loop
        if result is None or result.empty:
            print(f"Exiting loop.")
            break

        # Append the result to the list
        all_items.append(result)
        
        # Move to the next page
        current_page += 1

    # Print a completion message
    print(f"\nData retrieval completed successfully. Items retrieved {len(all_items)*items_per_page}")
    
    # Concatenate all collected data into a Pandas DataFrame and return
    return pd.concat(all_items, ignore_index=True)
```

```python
# Filter Populated Cities and Count Companies

def filter_populated_cities(df):
    
    df_active = df[df['company_status'] == 'active'].copy()

    # Custom function to extract locality
    def extract_locality(address):
        return address.get('locality', None)

    # Apply the function to create a new column 'locality'
    df_active['locality'] = df_active['registered_office_address'].apply(extract_locality)

    # Create a list of most populated cities in UK 2023 in the descending order of thier population
    cities_to_filter = ['London', 'Birmingham', 'Liverpool', 'Nottingham', 'Sheffield', 'Bristol', 'Glasgow', 'Leicester', 'Edinburgh', 'Leeds']

    # Create a dictionary with value counts
    dict_count = df_active['locality'].value_counts().to_dict()

    # Sort the dictionary based on the order of cities_to_filter
    dict_count_sorted = {city: dict_count.get(city, 0) for city in cities_to_filter}

    return dict_count_sorted
```

```python
# Plot for Locality Names
fig, ax = plt.subplots(figsize=(12, 10), subplot_kw={'projection': ccrs.PlateCarree()})
ax.coastlines()

# Scatter plot for wholesale counts
scatter_wholesale = ax.scatter(df_localities['longitude'], df_localities['latitude'], cmap='viridis', s=50, label='Wholesale')

# Annotate with count values
for idx, row in df_localities.iterrows():
    ax.annotate(f"{row['locality']}", (row['longitude'], row['latitude']), color='purple', fontsize=12, ha='right')

# Create a table with locality and count values
table_data = [['Locality', 'Count Growing', 'Count Wholesale', 'Count Retail', 'Count Preserving']] + df_localities[['locality', 'count_growing', 'count_wholesale', 'count_retail', 'count_preserving']].values.tolist()
table = plt.table(cellText=table_data, colLabels=None, cellLoc='center', bbox=[1.2, 0, 1.4, 1], fontsize=18)  # Adjust bbox parameter


# Apply conditional formatting to the table
for i in range(1, len(df_localities) + 1):
    count_growing = df_localities.loc[i - 1, 'count_growing']
    count_wholesale = df_localities.loc[i - 1, 'count_wholesale']
    count_retail = df_localities.loc[i - 1, 'count_retail']
    count_preserving = df_localities.loc[i - 1, 'count_preserving']

    # Highlight counts in red if less than the specified threshold, and in yellow if between the specified range
    def highlight_cell(cell, count, threshold, range_low, range_high, color):
        if 0 <= count < threshold:
            cell.set_facecolor(color)
            cell.set_text_props(color='white')
        elif range_low <= count <= range_high:
            cell.set_facecolor('yellow')

    highlight_cell(table[(i, 1)], count_growing, 10, 11, 20, 'red')  # Count Growing
    highlight_cell(table[(i, 2)], count_wholesale, 10, 11, 20, 'red')  # Count Wholesale
    highlight_cell(table[(i, 3)], count_retail, 10, 11,20 , 'red')  # Count Retail
    highlight_cell(table[(i, 4)], count_preserving, 5, 1, 4, 'red')  # Count Preserving

# Remove axes for the table
ax.axis('off')

plt.show()
```
